```python
import matplotlib.pyplot as plt #mengimport library matpolib
df['Date'] = pd.to_datetime(df['Date']) #menggubah format kolom Date menjadi format tanggal
plt.plot(df['Date'], df['Rainfall'])#visualisasi dengan axis horizontal untuk date dan axis vertikal untuk rentang nilai
```

```python
plt.plot(df['Date'], df['Rainfall'], label='Rainfall') #menampilkan data menjadi garis linear dalam grafik
plt.plot(df['Date'], df['Evaporation'], label='Evaporation')
plt.plot(df['Date'], df['Sunshine'], label='Sunshine')

plt.xlabel('Tahun') #judul axis horizontal
plt.ylabel('Rentang nilai') #judul axis vertikal
plt.title('Data Detail Cuaca di AUS per tahun by: IfanyDT') #judul grafik
plt.legend() #infromasi untuk warna garis
plt.grid(True) #informasi grid agar data mudah dibaca
```
